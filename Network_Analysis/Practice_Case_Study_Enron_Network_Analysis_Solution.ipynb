```python
import pandas as pd

import matplotlib.pyplot as plt

from decorator import decorator

import networkx as nx

from networkx.utils import create_random_state, create_py_random_state
```

```python
data = pd.read_csv('EmailEnron.csv')
```

```python
data.head()
```

```python
G = nx.Graph()
```

```python
G = nx.from_pandas_edgelist(data, 'From', 'To')

plt.figure(figsize = (10, 10))

options = {
    "node_color": "black",
    "node_size": 10,
    "linewidths": 0.5,
    "width": 0.1,
}

nx.draw(G, with_labels = True, **options)

plt.show()
```

```python
plt.figure(figsize = (10, 10))

nx.draw_shell(G, with_labels = True, **options)
```

```python
plt.figure(figsize = (10, 10))

# With the default parameters
nx.draw_spring(G, with_labels = True)
```

```python
#print(G)

# Let us quickly look at the degree of the nodes
for i in G.degree():
   print(i)
```

```python
# CENTRALITY MEASURES

deg_cen = nx.degree_centrality(G)

eig_cen = nx.eigenvector_centrality(G)

clo_cen = nx.closeness_centrality(G)

bet_cen = nx.betweenness_centrality(G)

#print(deg_cen, '\n')
```

```python
# DEGREE CENTRALITY

sort_deg_cen = sorted(deg_cen, key = deg_cen.get, reverse = True)
print('Nodes by decreasing degree centrality: \n',sort_deg_cen, '\n')

temp = {}

for w in sort_deg_cen:
#    print('w is: ', w)
    temp[w] = deg_cen[w]

print("Top 5 nodes in terms of degree centrality: {}".format(sort_deg_cen[:5]))
```

```python
list = sort_deg_cen[:5]

color = []

for node in G:
    if node in list:
        color.append('red')
    else:
        color.append('blue')
plt.figure(figsize = (10,10))
nx.draw(G, node_color =  color, with_labels = True)
```

```python
# EIGENVECTOR CENTRALITY

sort_eig_cen = sorted(eig_cen, key = eig_cen.get, reverse = True)
print('Nodes by decreasing eigenvector centrality: \n',sort_eig_cen, '\n')

#temp = {}

#for w in sort_eig_cen:
#    print('w is: ', w)
   # temp[w] = eig_cen[w]

print("Top 5 nodes in terms of eigenvector centrality: {}".format(sort_eig_cen[:5]))

```

```python
list = sort_eig_cen[:5]

color = []

for node in G:
    if node in list:
        color.append('red')
    else:
        color.append('blue')
plt.figure(figsize = (15,10))
nx.draw(G, node_color =  color, with_labels = True)
```

```python
# BETWEENNESS CENTRALITY

sort_bet_cen = sorted(bet_cen, key = bet_cen.get, reverse = True)
print('Nodes by decreasing betweenness centrality: \n', sort_bet_cen, '\n')

#temp = {}

#for w in sort_eig_cen:
#    print('w is: ', w)
   # temp[w] = eig_cen[w]

print("Top 5 nodes in terms of betweenness centrality: {}".format(sort_bet_cen[:5]))


list = sort_bet_cen[:5]
color = []
for node in G:
    if node in list:
        color.append('red')
    else:
        color.append('blue')
plt.figure(figsize = (15,10))
nx.draw(G, node_color =  color, with_labels = True)
```
